```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Player': ['Antoine Semenyo', 'Jarrod Bowen', 'Ismaila Sarr', 'Kaoru Mitoma'],
    '2023/2024': [10, 22, 7, 7],
    '2024/2025': [16, 21, 14, 14],
    '2025/2026': [11, 7, 4, 3]
}

df = pd.DataFrame(data)

# Melt the dataframe
df_melted = pd.melt(df, id_vars='Player', var_name='Season', value_name='G/A')

# Create a figure with three subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
seasons = ['2023/2024', '2024/2025', '2025/2026']

for i, season in enumerate(seasons):
    season_data = df_melted[df_melted['Season'] == season]
    sns.lineplot(data=season_data, x='Player', y='G/A', ax=axs[i], marker='o')
    axs[i].set_title(season)
    axs[i].set_xlabel('Player')
    axs[i].set_ylabel('Goals + Assists')
    axs[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Player': ['Antoine Semenyo', 'Jarrod Bowen', 'Ismaila Sarr', 'Kaoru Mitoma'],
    '2023/2024': [10, 22, 7, 7],
    '2024/2025': [16, 21, 14, 14],
    '2025/2026': [11, 7, 4, 3]
}

df = pd.DataFrame(data)

# Melt the dataframe
df_melted = pd.melt(df, id_vars='Player', var_name='Season', value_name='G/A')

# Create a line plot
plt.figure(figsize=(9, 4))
sns.lineplot(data=df_melted, x='Player', y='G/A', hue='Season', marker='o')
plt.title('Premier League Goals + Assists by Season')
plt.xlabel('Player')
plt.ylabel('Goals + Assists')
plt.tick_params(axis='x', rotation=45)
plt.show()
```
